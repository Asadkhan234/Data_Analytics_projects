# Vectorization
- Make The Data In The Token Form :

```python
# Bag of Words 
# Importing Library :
from sklearn.feature_extraction.text import CountVectorizer

# Here We Have Making A Variable Of cv In That We Store 5000 Words Which Is Relates To Each Others 
# Reomiving The Stop Words EX :(The , Puctuations , is , We etc......)
# Converting That 5000 Words Into Binary : (0,1) Form.
cv= CountVectorizer(max_features=5000, stop_words='english',binary=True)
```

```python
vector = cv.fit_transform(mobile['description']).toarray()
```

# Stemming

    It Works 2 Things :
    One Is It Can Remove All The Stop Words Which Is Present In The Nltk Library For Making The Array Short.
    Two Is I Can Change Multiple Words Occurs Into One Word .


```python
# This Is A Library 
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stopwords.words('english')
```

```python
# HERE WE ARE REMOVING STOP WORDS.
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#Function Of Steming :
#Spit The Text Input And Append In Y A Stem Text .
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
# CHEACKING STEM FUNCTION .
stem('''Apple iPhone SE (2022): Powered by iOS, with 4GB RAM and 128GB storage. Features a 12MP main camera and 7MP selfie camera. Battery
      2018mAh, 4.7-inch display, weighing 144g.Performance score 7.23/10. Released on 18/03/2022, priced at $429.''')
```

```python
# Apply stem safely
mobile = mobile.copy()  
mobile['description'] = mobile['description'].apply(stem)

# Access the 4th row (row index 3 by position)
print(mobile['description'].iloc[3])
```

# Cousine Similarity Check
- CHEACKING THE SIMILARITY USING NEAREST DEUPLICATED WORDS.

```python
# Here We Finding Cosine Similarity :
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(mob):
  Mobile_index = mobile[(mobile['brand']==mob)|(mobile['model']==mob)].index[0]
  distances = similarity[Mobile_index]
  mob_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x: x[1])[1:6]
  for i in mob_list:
      print(mobile.iloc[i[0]]['brand'])
      print(mobile.iloc[i[0]]['model'])
```

```python
recommend('Apple')
```
