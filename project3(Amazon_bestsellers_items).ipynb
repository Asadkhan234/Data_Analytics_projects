```python

# Calculate Q1, Q3 and IQR
Q1 = data['product_price'].quantile(0.25)
Q3 = data['product_price'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers (1.5 times IQR above Q3)
outliers = data[data['product_price'] > (Q3 + 1.5 * IQR)]

print(outliers[['product_title', 'product_price']])
```

```python
x = outliers.index

# y-axis: only the product_price column
y = outliers['product_price']

plt.figure(figsize=(10,5))
plt.plot(x, y, marker='o', linestyle='', color='red')  # scatter style
plt.xlabel("Product Index")
plt.ylabel("Price")
plt.title("Price Outliers")
plt.show()
```
