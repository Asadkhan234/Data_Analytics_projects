```python
plt.figure(figsize=(8,6))
plt.scatter(data['Profit'], data['Sales'], alpha=0.5)
plt.xlabel("Profit")
plt.ylabel("Sales")
plt.title("Sales vs Profit Scatter Plot")
plt.show()
```

MAX PROFIT PER SALE IS 8000

```python
city_count = data.groupby('City')['Sales'].count().sort_values(ascending=False).head(10)  # top 10 cities

plt.figure(figsize=(8,8))
plt.pie( city_count.values,labels=city_count.index, autopct='%.2f%%',  startangle=90)
plt.title("Order Count by Top 10 Cities")
plt.show()
```

MAX ORDERS FROM CITIES ARE NEW YORK , LOS ANGELOS AND PHILADELPHIA. RESPECTIVELY

```python
su = data.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)  # top 10 cities

plt.figure(figsize=(8,8))
plt.pie(su.values, labels=su.index, autopct='%1.1f%%', startangle=90)
plt.title("Top 10 Cities by Sales")
plt.show()
```

MAX SALES OR MONEY SPEND FROM CITIES ARE NEW YORK , LOS ANGELOS AND MANILA.

```python

country_summary = data.groupby('Country').agg({
    'Product Name': 'nunique',  # unique products sold
    'Quantity': 'sum'           # total quantity sold
})


plt.figure(figsize=(25,6))
plt.bar(country_summary.index, country_summary['Quantity'], color="skyblue")
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Total Quantity Sold")
plt.title("Total Quantity Sold by Country")
plt.show()
```

MAX QUANTITY SOLD FORM TOP 3 COUNTRIES ARE : UNITED STATES , FRANCE ,MEXIO  AUSTRIA.

```python
ax = data.groupby(['Order Priority','Order Date'])['Sales'].sum().reset_index()

plt.figure(figsize=(12,6))
for priority in ax['Order Priority'].unique():
    subset = ax[ax['Order Priority'] == priority]
    plt.plot(subset['Order Date'], subset['Sales'], marker='o', label=priority)



plt.legend()
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.title("Sales Trend by Order Priority")
plt.show()
```

MAX SALES COMES FROM MEDIUM SALES TREND.

```python
customer_sales = data.groupby('Customer Name')['Sales'].sum().reset_index()

# Filter customers with sales > 24000
high_value_customers = customer_sales[customer_sales['Sales'] > 24000]

fig, ax = plt.subplots(figsize=(12,6))

bars = ax.bar(high_value_customers['Customer Name'], high_value_customers['Sales'], color="skyblue")

# Add labels on top of bars
ax.bar_label(bars, padding=3, fontsize=9, rotation=90)

ax.set_xlabel("Customer Name")
ax.set_ylabel("Total Sales")
ax.set_title("Customers with Sales > 24000")
plt.xticks(rotation=90)
plt.show()
```

CUSTOMER NAMES WITH SALES  GREATER THAN 24000.
