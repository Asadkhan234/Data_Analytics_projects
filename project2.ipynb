# Data Visuallization : In bar Plot

```python
import matplotlib.pyplot as plt
```

```python
Mared_Unmareed['Label'] = Mared_Unmareed['Gender'] + " - " + Mared_Unmareed['ParentMaritalStatus']

plt.bar(Mared_Unmareed['Label'], Mared_Unmareed['Count'])
plt.xticks(rotation=45)
plt.xlabel("Gender - Parent Marital Status")
plt.ylabel("Count")
plt.title("Parent Marital Status by Gender")
plt.show()
```

```python
plt.bar(
    Edu_Grp['EthnicGroup'] + "_" + Edu_Grp['Gender'] + "_" + Edu_Grp['ParentMaritalStatus'], 
    Edu_Grp[('ParentEduc','count')]
)

plt.xticks(rotation=90)
plt.xlabel("EthnicGroup - Gender - ParentMaritalStatus")
plt.ylabel("Count")
plt.title("Parent Education Count by Groups")
plt.show()
```

```python
Chield['Group'] = Chield['Gender'] + " | " + Chield['ParentMaritalStatus'] + " | Siblings: " + Chield['NrSiblings'].astype(str)

# Plot
plt.figure(figsize=(10,6))
plt.bar(Chield['Group'], Chield[('IsFirstChild', 'count')])

# Beautify
plt.xticks(rotation=90)
plt.xlabel("Gender | ParentMaritalStatus | NrSiblings")
plt.ylabel("Count of IsFirstChild")
plt.title("Count of First Child by Gender, Parent Status, and Siblings")
plt.tight_layout()
plt.show()
```

```python
plt.bar(Trans['WklyStudyHours'], Trans['TransportMeans']['count'])
plt.xlabel("Weekly Study Hours")
plt.ylabel("Count of TransportMeans")
plt.title("Transport Means vs Weekly Study Hours")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(score['ParentEduc'], score['MathScore'], label='Math')
plt.bar(score['ParentEduc'], score['ReadingScore'], bottom=score['MathScore'], label='Reading')
plt.bar(score['ParentEduc'], score['WritingScore'], 
        bottom=score['MathScore']+score['ReadingScore'], label='Writing')

plt.xticks(rotation=45)
plt.xlabel("Parent Education Level")
plt.ylabel("Scores")
plt.title("First Scores by Parent Education Level")
plt.legend()
plt.show()
```

```python
plt.bar(Prep['TestPrep'], Prep['Gender'])  
plt.xlabel("Test Preparation")  
plt.ylabel("Count of Students")  
plt.title("Students Count by Test Preparation")  
plt.show()
```

```python
plt.bar(c['Gender'], c['PracticeSport']['count'],color=['red','green'])
plt.xlabel("Gender")
plt.ylabel("Count of Students")
plt.title("Students Practicing Sport by Gender")
plt.show()
```

```python
plt.bar(lunh['Gender'], lunh['LunchType']['count'])
plt.xlabel("Gender")
plt.ylabel("Count of LunchType")
plt.title("Lunch Type Count by Gender")
plt.show()
```
