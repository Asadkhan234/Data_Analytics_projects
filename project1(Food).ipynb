```python
data['Fat'] = data['Fat'].round(2)
data['Fibre'] = data['Fibre'].round(2)

data['Protein'] = data['Protein'].round(2)
data['Carbs'] = data['Carbs'].round(2)
```

```python
data[['Protein','Carbs','Fat']] = np.where(data[['Protein','Carbs','Fat']]<=0.0,data[['Protein','Carbs','Fat']].mean(),data[['Protein','Carbs','Fat']])
```

```python
data[['Protein','Carbs']] = data[['Protein','Carbs']].round(2)
```

```python
import random 
foods = ["Rice", "Bread", "Oats", "Apple", "Banana", "Potato", "Pasta", "Nuts", "Milk", "Egg"]
data["Food"] = random.choices(foods, k=len(data))
data
```

```python
data["Year"] = np.random.randint(2015, 2026, size=len(data))
```
